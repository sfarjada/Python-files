Creating test and traning set for Predictive Models


```python
dropS = ['Unnamed: 0','Type of Control','Year', 'Rural versus Urban']
for i in dropS:
  df.drop(columns=[i], inplace=True)
```

```python
df.head()
```

```python
# creating test training set.
from sklearn.model_selection import train_test_split

# all columns, that is not 'IsGoodNh'
X = df.drop(columns='IsGoodNh')
# dependent variable 'IsGoodNh'
y = df['IsGoodNh']

# Create train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=5/30,
random_state=101)
```

```python
# reviewing split size
train_test_shapes = {
    'X_train': X_train.shape,
    'X_test': X_test.shape,
    'y_train': y_train.shape,
    'y_test': y_test.shape
}

train_test_shapes
```

```python
# Attempting a logistics Regression, and running a classification report
from sklearn.linear_model import LogisticRegression


# Running the logisitc Regression
log_reg = LogisticRegression(max_iter=1000, random_state=42)

# Fitting the model
log_reg.fit(X_train, y_train)

# Predicting the Model
y_pred = log_reg.predict(X_test)
```

```python
# Calculate coeeficient values
coefficients = log_reg.coef_[0]
coeff_series = pd.Series(data=coefficients, index=X_train.columns)

sorted_coeff_series = coeff_series.sort_values(ascending=False).round(2)

sorted_coeff_series
```

```python
# Create a bar graph checking to visualize the coeeficient values.
plt.figure(figsize=(12, 10))
sns.barplot(x=sorted_coeff_series.values, y=sorted_coeff_series.index, palette='coolwarm')
plt.title(' Logistic Regression Feature Coefficients')
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.grid(True)
plt.show()
```

```python
#logistic regression 2nd attempt to get predictions
log_reg = LogisticRegression(max_iter=1000, random_state=101)
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)

#Reviewing predictions
print(y_pred[:100])
```

```python
# running decision tree classifier
from sklearn.tree import DecisionTreeClassifier

# Running the model
clf_tree_original = DecisionTreeClassifier(max_depth=3)

# Fitting the model
clf_tree_original.fit(X_train, y_train)

# Prediction
y_pred_tree_original = clf_tree_original.predict(X_test)
```

```python
# Outputting the decision tree
from sklearn.tree import plot_tree

plt.figure(figsize = (20,10))
plot_tree(clf_tree_original, filled=True,rounded=True,class_names =['0','1'], feature_names=X_train.columns)
plt.title("Decision tree")
plt.show()
```

```python
# Calculating coefficients
cf_importance =pd.Series(data=clf_tree_original.feature_importances_, index=X_train.columns).sort_values(ascending=False).round(3)
cf_importance
```

```python
# Decision tree ceofficients plot.
plt.figure(figsize=(12, 12))
sns.barplot(x=cf_importance.values, y=cf_importance.index, palette='coolwarm')
plt.title('Classification Feature Coefficients')
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.grid(True)
plt.show()
```

```python
# Running Random forest

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42,n_estimators=99,
                            max_features=6,
                            max_depth=3,
                            min_samples_split=100)

# Fitting the model
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_train)
```

```python
#calculating coefficients.
rf_importance = pd.Series(data=rf.feature_importances_, index=X_train.columns).sort_values(ascending=False).round(3)
rf_importance
```

```python
# Visualizing coefficients.
plt.figure(figsize=(12, 12))
sns.barplot(x=rf_importance.values, y=rf_importance.index, palette='coolwarm')
plt.title(' Random Forest Feature Coefficients')
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.grid(True)
plt.show()
```

```python
# Calculating AUC and creating and ROC Curve
from sklearn.metrics import roc_auc_score, roc_curve, auc

y_prob_log_reg = log_reg.predict_proba(X_test)[:, 1]

y_prob_rf = rf.predict_proba(X_test)[:, 1]

y_prob_tree = clf_tree_original.predict_proba(X_test)[:, 1]

# Calculate AUC
auc_log_reg = roc_auc_score(y_test, y_prob_log_reg)
auc_rf = roc_auc_score(y_test, y_prob_rf)
auc_tree = roc_auc_score(y_test, y_prob_tree)

# Print AUC
print(f"AUC - Logistic Regression: {auc_log_reg:.3f}")
print(f"AUC - Random Forest: {auc_rf:.3f}")
print(f"AUC - Decision Tree: {auc_tree:.3f}")


# Logistic Regression ROC and plotting the ROC
plt.figure(figsize=(8, 6))
fpr_log, tpr_log, _ = roc_curve(y_test, y_prob_log_reg)
plt.plot(fpr_log, tpr_log, label=f'Logistic Regression (AUC = {auc_log_reg:.3f})')

# Random Forest ROC
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc_rf:.3f})')

# Decision Tree ROC
fpr_tree, tpr_tree, _ = roc_curve(y_test, y_prob_tree)
plt.plot(fpr_tree, tpr_tree, label=f'Decision Tree (AUC = {auc_tree:.3f})')

# Final ROC plot
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Logistic Regression, Random Forest, and Decision Tree')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Decision Tree Predictions and Accuracy
y_pred_tree = clf_tree_original.predict(X_test)
accuracy_tree = accuracy_score(y_test, y_pred_tree)
print(f"Decision Tree Accuracy: {accuracy_tree:.3f}")

# Logistic Regression Predictions and Accuracy
y_pred_log_reg = log_reg.predict(X_test)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
print(f"Logistic Regression Accuracy: {accuracy_log_reg:.3f}")

# Random Forest Predictions and Accuracy
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf:.3f}")
```

```python
accuracies = pd.DataFrame(columns=['train', 'test'], index=['LogisticReg', 'ClassTree', 'RF'])

model_dict = {'LogisticReg': log_reg, 'ClassTree': clf_tree_original, 'RF': rf}

# Calculate accuracy scores for each model
for name, model in model_dict.items():
    accuracies.loc[name, 'train'] = accuracy_score(y_train, model.predict(X_train))
    accuracies.loc[name, 'test'] = accuracy_score(y_test, model.predict(X_test))

# Display the accuracies DataFrame
accuracies
```

```python
# Create a figure and a single subplot
fig, ax = plt.subplots(figsize=(10, 5))

# Sort the DataFrame by the 'Accuracy' column in descending order and plot
accuracies.sort_values(by='test', ascending=False).plot(kind='barh', ax=ax, zorder=3)

# Set grid with zorder set to 0 to go behind the bars
ax.grid(zorder=0)

plt.title("Model Accuracy Comparison")
plt.xlabel("Accuracy Score")
plt.ylabel("Model")
plt.show()
```
