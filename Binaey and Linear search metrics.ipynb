```python
def main():
    try:
        numbers = input('Enter your numbers in order, seperate by a space: ')
        search = int(input("Enter your target number: "))

        seperate = numbers.split(' ')

        numberList = [int(element.strip()) for element in seperate if element.strip()]

        isSorted = True
        for i in range(len(numberList) - 1):
            if numberList[i] > numberList[i + 1]:
                isSorted = False
                break
        if isSorted == True:
          binarySearch,iterationb = binary(numberList,search)
          binarySearchR,iterationbr = binaryR(numberList,search)
        else:
          binarySearch = -1
          binarySearchR = -1
        linearSearch,iterationls = sequentialSearch(search, numberList)
        linearSearchR,iterationlsr = recursiveLinearSearch(search, numberList)

        if binarySearch !=-1 and binarySearchR !=-1 and linearSearch !=-1 and linearSearchR !=-1:
            print('Positon of the binary search iteratively: ',binarySearch)
            print('Positon of the binary search recursively: ',binarySearchR)
            print('Position of the linear search interatively: ',linearSearch)
            print('position of the linear search recusviely: ', linearSearchR)
        elif linearSearch !=-1 and linearSearchR !=-1:
            print('Position of the linear search interatively: ',linearSearch)
            print('position of the linear search recusviely: ', linearSearchR)
            print('List is not sorted cannot find number via binary methods')
        else:
            print("number cannot be found in the list")
    except ValueError:
        print("please enter a valid integer and space them out")


def binary(list,target):
    left = 0
    right = len(list)-1
    countb = 0

    while left <= right:
        countb=countb+1
        midpoint = (left + right) // 2
        if list[midpoint] == target:
            return midpoint, countb
        elif list[midpoint] > target:
            right = midpoint - 1
        else:
            left = midpoint + 1
    return - 1, countb

def binaryR(list ,target, left = 0, right=None,countbr=0):
    countbr=+1
    if right is None:
        right = len(list)-1

    midpoint = (left + right) // 2

    if left > right:
        return -1,countbr
    elif list[midpoint] > target:
        return binaryR(list,target, left, midpoint-1,countbr)
    elif list[midpoint] < target:
        return binaryR(list,target, midpoint+1, right,countbr)
    else:
        return midpoint,countbr

def sequentialSearch(target, lyst):
    position = 0
    countls = 0
    while position < len(lyst):
        countls=countls+1
        if target == lyst[position]:
            return position,countls
        position += 1
    return -1, countls

def recursiveLinearSearch(target, lyst, position=0,countlsr=0):
    countlsr=countlsr+1
    if position >= len(lyst):
        return -1,countlsr


    if lyst[position] == target:
        return position,countlsr


    return recursiveLinearSearch(target, lyst, position + 1,countlsr)
```

```python
numbers = [1, 3, 5, 7, 9, 10, 11, 13, 17, 29]
search = 11


isSorted = all(numbers[i] <= numbers[i + 1] for i in range(len(numbers) - 1))

if isSorted:
  start = time.time()
  binarySearch, binaryIterations = binary(numbers, search)
  binaryTime = time.time() - start

  start = time.time()
  binarySearchR, binaryRecursiveIterations = binaryR(numbers, search)
  binaryRecursiveTime = time.time() - start
else:
  binarySearch = binarySearchR = -1
  binaryIterations = binaryRecursiveIterations = 0
  binaryTime = binaryRecursiveTime = 0

start = time.time()
linearSearch, linearIterations = sequentialSearch(search, numbers)
linearTime = time.time() - start

start = time.time()
linearSearchR, linearRecursiveIterations = recursiveLinearSearch(search, numbers)
linearRecursiveTime = time.time() - start


print(f'Position of binary search (iteratively): {binarySearch}, Iterations: {binaryIterations}, Time: {binaryTime:.6f} seconds')
print(f'Position of binary search (recursively): {binarySearchR}, Iterations: {binaryRecursiveIterations}, Time: {binaryRecursiveTime:.6f} seconds')
print(f'Position of linear search (iteratively): {linearSearch}, Iterations: {linearIterations}, Time: {linearTime:.6f} seconds')
print(f'Position of linear search (recursively): {linearSearchR}, Iterations: {linearRecursiveIterations}, Time: {linearRecursiveTime:.6f} seconds')
```

```python
mem_usage_binary = max(memory_usage((binary, (numbers, search))))
mem_usage_binaryR = max(memory_usage((binaryR, (numbers, search))))
mem_usage_linear = max(memory_usage((sequentialSearch, (search, numbers))))
mem_usage_linearR = max(memory_usage((recursiveLinearSearch, (search, numbers))))
print(f'Memory usage (binary iterative): {mem_usage_binary:.2f} MiB')
print(f'Memory usage (binary Recursive): {mem_usage_binaryR:.2f} MiB')
print(f'Memory usage (linear iterative): {mem_usage_linear:.2f} MiB')
print(f'Memory usage (linear recursive): {mem_usage_linearR:.2f} MiB')
```
